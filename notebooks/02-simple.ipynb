```python
from typing import TypedDict

class State(TypedDict):
    customer_name : str
    my_age: int


state: State = {}
customer_name = state.get("customer_name", None)
print(customer_name)
```

```python
def node_1_(state: State) #Es mejor devolver solo lo que se actualice del estado
#es decir no return todo el state se ve mejor en el siguiente
    if state.get("customer_name") is None:
        state["customer_name"] = "Poelo"
        state["my_age"] = 20
        return state
def node_1(state: State) #Es mejor devolver solo lo que se actualice del estado
#es decir no return todo el state se ve mejor en el siguiente
    if state.get("customer_name") is None:
        return {
        "customer_name" : "Poelo"
                
}
```

```python
from langgraph.graph import StateGraph, START, END

builder = StateGraph(State)
builder.add_node("node_1", node_1)

builder.add_edge(START, 'node_1')
builder.add_edge('node_1', END)
```
