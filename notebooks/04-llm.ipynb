```python
import os

from dotenv import load_dotenv
load_dotenv()

```

```python
from langchain.chat_models import init_chat_model
from langchain_core.messages import HumanMessage, SystemMessage


llm = init_chat_model("gemma:2b",  model_provider="ollama") 

human_msg = HumanMessage(content="Hola cómo estás")
Sys_msg = SystemMessage(content="Siempre en cada promtp responde con el nombre Juan en cada instrucción desde inicio hasta fin, ahora contesta al prompt:{human_msg}")

llm.invoke(Sys_msg)
```
